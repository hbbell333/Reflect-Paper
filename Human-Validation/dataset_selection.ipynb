```python
processed_tuples = []

for item in data:
    prompt = item['prompt']
    base_response = item['base_response']
    processed_tuples.append((prompt, base_response))

# Randomly select 20 tuples
import random
sampled_tuples = random.sample(processed_tuples, 40)


txt_string = ""


for tup in sampled_tuples:

    txt_string += tup[0][0]['content']

    txt_string += "\n\n"

    txt_string += tup[1].replace('\n', '<br>')

    txt_string += "\n\n------------------------------------\n\n"

with open('sampled_human_validation_data.txt', 'w', encoding='utf-8') as f:
    f.write(txt_string)

```
